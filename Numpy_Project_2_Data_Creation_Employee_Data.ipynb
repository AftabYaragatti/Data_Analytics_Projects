##### **Project Objective :** Analyze employee data and answer business questions such as:
##### => Highest paid employee
##### => Lowest paid employee
##### => Average salary
##### => Department-wise analysis
##### => Bonus calculation
##### => Salary hike
##### => Employee filtering
##### => Performance analysis

##### **Step 1:** Import NumPy

```python
import numpy as np
```

##### **Step 2:** Create Employee Dataset
##### Employee IDs

```python
emp_id = np.array([101,102,103,104,105,106,107,108])
```

##### Employee Names

```python
emp_name = np.array([
    "Aftab",
    "Rahul",
    "Priya",
    "Sneha",
    "Arjun",
    "Kiran",
    "Neha",
    "Rohan"
])
emp_name
```

##### Departments

```python
department = np.array([
    "IT",
    "HR",
    "Finance",
    "IT",
    "Sales",
    "HR",
    "Finance",
    "IT"
])
department
```

##### Salary

```python
salary = np.array([
    50000,
    45000,
    70000,
    55000,
    40000,
    48000,
    75000,
    60000
])
salary
```

##### Experience

```python
experience = np.array([
    2,
    1,
    5,
    3,
    1,
    2,
    6,
    4
])
experience
```

##### Performance Rating

```python
rating = np.array([
    4,
    3,
    5,
    4,
    2,
    3,
    5,
    4
])
rating
```

##### **Step 3:** Display Complete Employee Data

```python
for i in range(len(emp_id)):
    print(emp_id[i],
          emp_name[i],
          department[i],
          salary[i],
          experience[i],
          rating[i])
```

##### **Step 4:** Total Number of Employees

```python
print(len(emp_id))
```

##### **Step 5:** Average Salary

```python
print(np.mean(salary))
```

##### **Step 6:** Highest Salary

```python
max_salary = np.max(salary)
print(max_salary)
```

##### **Step 7:** Employee with Highest Salary

```python
index = np.argmax(salary)
print(emp_name[index])
print(salary[index])
```

##### **Step 8:** Lowest Salary Employee

```python
index = np.argmin(salary)
print(emp_name[index])
print(salary[index])
```

##### **Step 9:** Employees Earning More Than 50,000

```python
print(emp_name[salary > 50000])
```

##### **Step 10:** Employees in IT Department

```python
print(emp_name[department == "IT"])
```

##### **Step 11:** Employees Having Experience More Than 3 Years

```python
print(emp_name[experience > 3])
```

##### **Step 12:** Add 10% Salary Hike

```python
new_salary = salary * 1.10
print(new_salary)
```

##### **Step 13:** Bonus Based on Performance
##### Rating 5 → ₹10000 Bonus
##### Rating 4 → ₹5000 Bonus
##### Others → ₹2000 Bonus

```python
bonus = np.where(
    rating == 5,
    10000,
    np.where(
        rating == 4,
        5000,
        2000
    )
)
print(bonus)
```

##### **Step 14:** Final Salary

```python
final_salary = salary + bonus
print(final_salary)
```

##### **Step 15:** Top Performers

```python
print(emp_name[rating == 5])
```

##### **Step 16:** Employees Eligible for Promotion
##### Conditions:
##### Experience ≥ 3
##### Rating ≥ 4

```python
promotion = emp_name[ (experience >= 3) & (rating >= 4)]
print(promotion)
```

##### **Step 17:** Department-wise Employee Count

```python
unique_dept = np.unique(department)
for dept in unique_dept:
    count = np.sum(department == dept)
    print(dept, count)
```

##### **Step 18:** Department-wise Average Salary

```python
for dept in np.unique(department):
    avg_salary = np.mean(
        salary[department == dept]
    )
    print(dept, avg_salary)
```

##### **Step 19:** Sort Employees by Salary
##### Ascending

```python
index = np.argsort(salary)
print(emp_name[index])
```

##### Descending

```python
index = np.argsort(salary)[::-1]
print(emp_name[index])
```

##### **Step 20:** Statistical Analysis

```python
print(np.mean(salary))
print(np.median(salary))
print(np.std(salary))
print(np.var(salary))
```

##### **Step 21:** Employee Search by ID

```python
search_id = 104
index = np.where(emp_id == search_id)
print(emp_name[index])
```

##### **Step 22:** Generate Employee Report

```python
for i in range(len(emp_id)):
    print(
        f"""
Employee ID : {emp_id[i]}
Employee Name : {emp_name[i]}
Department : {department[i]}
Salary : {salary[i]}
Experience : {experience[i]}
Rating : {rating[i]}
Bonus : {bonus[i]}
Final Salary : {final_salary[i]}
----------------------------------
"""
    )
```

```python
emp_name[np.argmax(salary)]
```

##### Q2. How many employees earn above average salary?

```python
np.sum(salary > np.mean(salary))
```

##### Q3. Which department has highest average salary?

```python
unique_dept = np.unique(department)
avg_salaries = [
    np.mean(salary[department == dept])
    for dept in unique_dept
]
print(unique_dept[np.argmax(avg_salaries)])
```
