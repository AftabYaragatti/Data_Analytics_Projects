##### Still Number columns have some string data type so we need to convert string data type to Integer by replacing Fires to Null string.

```python
df['NUMBER'].unique()
```

##### Extract the numeric part (digits and optional decimal points) Remove 'Fires' and strip extra spaces

```python
df['NUMBER'] = df['NUMBER'].str.replace('Fires', '', case=False).str.strip()
```

##### For Column Number change the Data type object to integer.

```python
df['NUMBER'].unique()
```

##### Rename 'NUMBER' to 'FIRES' and apply changes to the original df

```python
df.rename(columns={'NUMBER': 'FIRES'}, inplace=True)
```

```python
df.sample(3)
```

##### Conver the Fires column data type to Integer. Round to 0 decimal places, then convert 

```python
df['FIRES'].isna().sum()
```

```python
df['FIRES'].dtype
```

```python
df['FIRES'].replace(np.nan,0,inplace=True)
```

```python
# 1. Convert strings (like '1.976') to floats first
df['FIRES'] = pd.to_numeric(df['FIRES'], errors='coerce')

# Converts 1.976 to 2
df['FIRES'] = df['FIRES'].round(0).astype('Int64')
```
