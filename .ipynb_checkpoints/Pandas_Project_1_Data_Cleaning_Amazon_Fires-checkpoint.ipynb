## **PANDAS PROJECT 1 : Data Cleaning on Amazon Fires Dataset**

##### Import Neccessary Libraries

```python
import pandas as pd
import numpy as np
```

##### Read the Amazon Fires CSV File

```python
df = pd.read_csv("amazon_fires (1).csv",encoding='ISO-8859-4')
```

```python
df
```

#### **Exploratory Data Analysis**

##### Get the Unique values of Mes column to understand the data

```python
df['mes'].unique()
```

##### Since it is Spanish Language , Mes means Months we need to change the columns name mes to month

```python
new_columns = {'ano':'Year','mes':'Month','estado':'State','numero':'Number','encontro':'Meeting_Date'}
```

```python
new_columns
```

##### changing the column names

```python
df.rename(columns=new_columns,inplace=True)
```

```python
df.head(3)
```

##### Check that any column contains any null values or not. If yes replace or fill by other value.

```python
df.info()
```

##### To check the total Number of duplicates we need to use duplicated.

```python
df.duplicated().sum()
```

##### To drop the duplicates we have to use drop_duplicated.

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

##### again check whether we have duplicate and null values or not.6454 entries to 6424 entries totaly 30 duplicate records are deleted.

```python
df.info()
```

```python
df.head(3)
```

##### Get the unique values of Month table to rename the month present in spanish to convert it into english.

```python
df['Month'].unique()
```

##### We have to rename all the spanish months to English months name.

```python
new_month = {'Janeiro':'Jan', 'Fevereiro':'Feb', 'MarÄ¯o':'mar', 'Abril':'apr', 'Maio':'may', 'Junho':'June', 'Julho':'July',
       'Agosto':'aug', 'Setembro':'September', 'Outubro':'oct', 'Novembro':'Nov', 'Dezembro':'Dec'}
```

```python
new_month
```

##### See it is not changed the Months name to change it we need use map() method to map new_month to month column

```python
df['Month']
```

```python
df['Month'] = df['Month'].map(new_month)
```

```python
df['Month'] 
```

```python
df.sample(5)
```

##### Again Here data is not present in same case alphabets and Not is proper format to make it proper format we have to replace all months with first letter should be upper case followed by lowercase and Should be same format . Like we should convert jan -> January or January-> Jan

```python
df['Month'].unique()
```

```python
df['Month'].replace('September','sept',inplace=True)
```

```python
df['Month'].replace({'mar':'Mar','apr':'Apr'},inplace=True)
```

```python
df['Month'].unique()
```

##### Convert all months to the Uppercase.

```python
df['Month'] = df['Month'].str.upper()
```

```python
df['Month'].unique()
```

```python
df.sample(5)
```

##### Convert all columns names to UpperCase

```python
df.columns = df.columns.str.upper()
```

```python
df.sample(3)
```

##### Convert the States data to capitalise()

```python
df['STATE']=df['STATE'].str.capitalize()
```

```python
df['STATE']
```

```python
df.sample(3)
```

```python
df['NUMBER'].unique()
```

```python
df['NUMBER'].isna().sum()
```

```python
df[df['NUMBER'].isna()]
```

```python
df['NUMBER'] = df['NUMBER'].replace(np.nan,0)
```

```python
df['NUMBER'].isna().sum()
```

```python
df.sample(5)
```

```python
df['NUMBER'].unique()
```

```python
# 1. Extract the numeric part (digits and optional decimal points)
# Remove 'Fires' and strip extra spaces
df['NUMBER'] = df['NUMBER'].str.replace('Fires', '', case=False).str.strip()
```

```python
# 2. Convert to numeric (float) first to handle the decimals
df['NUMBER'] = pd.to_numeric(df['NUMBER'], errors='coerce')
```

##### For Column Number change the Data type object to integer.

```python
df['NUMBER'].unique()
```

```python
# Rename 'NUMBER' to 'FIRES' and apply changes to the original df
df.rename(columns={'NUMBER': 'FIRES'}, inplace=True)
```

```python
df.sample(3)
```

```python
df['NUMBER']=df['NUMBER'].astype(int)
```

```python
df['NUMBER'].dtype
```

```python
df.info()
```

```python
df['MEETING_DATE']
```

```python
# Replaces the original column with the datetime type
df['MEETING_DATE'] = pd.to_datetime(df['MEETING_DATE'])
```

```python
df.info()
```

```python
df.sample(7)
```

```python
df.to_csv("CLEANED_AMAZON_FIRES_DATASET.csv", index=False)
```
