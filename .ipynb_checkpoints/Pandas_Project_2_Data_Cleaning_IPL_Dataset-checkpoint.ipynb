####  **Level 2: Team Analysis**

##### 8. Which team has won the most matches?

```python
df['winner'].value_counts()
```

```python
df['winner'].value_counts().idxmax()
```

##### 9. Which team has played the most matches?

```python
pd.concat([df['team1'], df['team2']]).value_counts()
```

```python
pd.concat([df['team1'], df['team2']]).value_counts().idxmax()
```

##### 10. What is the win percentage of each team?

```python
matches_played = pd.concat([df['team1'], df['team2']]).value_counts()
matches_won = df['winner'].value_counts()
win_percentage = (matches_won / matches_played) * 100
win_percentage.sort_values(ascending=False)
```

##### 11. Which team has the highest win percentage (min 50 matches)?

```python
valid_teams = matches_played[matches_played >= 50]
win_percentage_filtered = (matches_won[valid_teams.index] / valid_teams) * 100
win_percentage_filtered.sort_values(ascending=False)
```

```python
win_percentage_filtered.idxmax()
```

##### 12. Which team has won the most matches in a single season?

```python
df.groupby(['season', 'winner']).size().sort_values(ascending=False)
```

```python
df.groupby(['season', 'winner']).size().sort_values(ascending=False).idxmax()
```

##### 13. Which team wins more when batting first vs chasing?

```python
df['win_type'] = df['result'].apply(lambda x: 'batting_first' if x == 'runs' else 'chasing')
```

```python
df.groupby(['winner', 'win_type']).size().unstack()
```

```python
df.groupby(['winner', 'win_type']).size().unstack().idxmax()
```

#### **Level 3: Toss Analysis**

##### 14. How many times did the toss winner also win the match?

```python
(df['toss_winner'] == df['winner']).sum()
```

```python
(df['toss_winner'] == df['winner']).sum().item()
```

##### 15. What is the toss win → match win percentage?

```python
toss_match_win = (df['toss_winner'] == df['winner']).sum()
total_matches = df.shape[0]
percentage = (toss_match_win / total_matches) * 100
percentage
```

```python
percentage.item()
```

##### 16. Which team benefits the most from winning the toss?

```python
toss_win_matches = df[df['toss_winner'] == df['winner']]
toss_win_matches['winner'].value_counts()
```

```python
toss_win_matches['winner'].value_counts().idxmax()
```

##### 17. What is the most common toss decision (`bat` or `field`)?

```python
df['toss_decision'].value_counts()
```

```python
df['toss_decision'].value_counts().idxmax()
```

##### 18. Does choosing to bat first increase chances of winning?

```python
bat_first = df[df['toss_decision'] == 'Batting']
bat_win = (bat_first['toss_winner'] == bat_first['winner']).sum()
bat_total = bat_first.shape[0]
bat_win_percentage = (bat_win / bat_total) * 100
bat_win_percentage
```

```python
bat_win_percentage.item()
```

```python
field_first = df[df['toss_decision'] == 'Fielding']
field_win = (field_first['toss_winner'] == field_first['winner']).sum()
field_total = field_first.shape[0]
field_win_percentage = (field_win / field_total) * 100
field_win_percentage
```

```python
field_win_percentage.item()
```

####  **Level 5: Venue & City Insights**

##### 23. Which venue hosted the most matches?

```python
df.groupby(['venue', 'winner']).size().reset_index(name='wins')
```

##### 25. Which city has the highest average result margin?

```python
df.groupby('city')['result_margin'].mean().sort_values(ascending=False)
```

##### 26. Which venue favors chasing teams more?

```python
df['win_type'] = df['result'].apply(lambda x: 'batting_first' if x == 'runs' else 'chasing')
df.groupby(['venue', 'win_type']).size().unstack()
```

```python
venue_stats = df.groupby(['venue', 'win_type']).size().unstack()
venue_stats['chasing_%'] = (venue_stats['chasing'] / venue_stats.sum(axis=1)) * 100
venue_stats.sort_values('chasing_%', ascending=False)
```
