```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ipl=pd.read_csv('CLEANED_IPL_Match_Dataset.csv')
```

```python
ipl
```

##### 1. Create a line plot showing how many matches where played in each season

```python
total_matches= ipl['season'].value_counts().sort_index()
```

```python
total_matches
```

```python
ipl['season'].unique()
```

```python
plt.plot(ipl['season'].unique(),total_matches,marker='o',markerfacecolor='red')
plt.title('Season Vs Total Number of Matches',color='blue')
plt.xlabel('Seasons',color='red')
plt.ylabel('Total Number of Matches',color='red')
```

##### 2. Plot the average result_margin for each season using a line graph

```python
avg_res=ipl.groupby(ipl['season'])['result_margin'].mean()
```

```python
avg_res
```

```python
plt.plot(ipl['season'].unique(),avg_res,marker='o',markerfacecolor='red')
plt.title('Season Vs Average Result Margin',color='blue')
plt.xlabel('Seasons',color='red')
plt.ylabel('Result Margin',color='red')
```

##### 3. Create a line plot showing the count of matches with super_over == 'Yes' for every season

```python
super_matches=ipl[ipl['super_over']=='Yes']
super_matches
```

```python
count_of_matches=super_matches['season'].value_counts().sort_index()
count_of_matches
```

```python
plt.plot(count_of_matches.index, count_of_matches.values, marker='o', markerfacecolor='red')
plt.title('Season Vs Number of Super Over Matches', color='blue')
plt.xlabel('Seasons', color='red')
plt.ylabel('Number of Super Over Matches', color='red')
plt.show()
```

##### 4. Create a line plot  comparing number of times teams choose batting but number of times teams choose field.

```python
bat_decision = (
    ipl[ipl['toss_decision'] == 'Batting']
    .groupby('season')['id']
    .count()
)
field_decision = (
    ipl[ipl['toss_decision'] == 'Fielding']
    .groupby('season')['id']
    .count()
)
plt.figure(figsize=(10,5))
plt.plot(bat_decision.index, bat_decision.values,
         marker='o', label='Bat First')
plt.plot(field_decision.index, field_decision.values,
         marker='o', label='Field First')
plt.title('Toss Decision Trend Over Seasons')
plt.xlabel('Season')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.show()
```

##### 5. Create a team (example Mumbai Indians or CSK) and plot the number of matches won by the team in each season

```python
team_name = 'Mumbai Indians'
team_wins = (
    ipl[ipl['winner'] == team_name]
    .groupby('season')['id']
    .count()
)
plt.figure(figsize=(10,5))
plt.plot(team_wins.index, team_wins.values, marker='o')
plt.title(f'{team_name} Wins Per Season')
plt.xlabel('Season')
plt.ylabel('Number of Wins')
plt.grid(True)
plt.show()
```

### subplot()

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,12))

# ====================== 1st Figure ======================
total_matches = ipl['season'].value_counts().sort_index()

plt.subplot(3,2,1)
plt.plot(total_matches.index,
         total_matches.values,
         marker='o',
         markerfacecolor='red')

plt.title('Season Vs Total Number of Matches', color='blue')
plt.xlabel('Seasons', color='red')
plt.ylabel('Total Number of Matches', color='red')
plt.grid(True)


# ====================== 2nd Figure ======================
avg_res = ipl.groupby('season')['result_margin'].mean()

plt.subplot(3,2,2)
plt.plot(avg_res.index,
         avg_res.values,
         marker='o',
         markerfacecolor='red')

plt.title('Season Vs Average Result Margin', color='blue')
plt.xlabel('Seasons', color='red')
plt.ylabel('Result Margin', color='red')
plt.grid(True)


# ====================== 3rd Figure ======================
super_matches = ipl[ipl['super_over'] == 'Yes']

count_of_matches = (
    super_matches['season']
    .value_counts()
    .sort_index()
)

plt.subplot(3,2,3)
plt.plot(count_of_matches.index,
         count_of_matches.values,
         marker='o',
         markerfacecolor='red')

plt.title('Season Vs Number of Super Over Matches', color='blue')
plt.xlabel('Seasons', color='red')
plt.ylabel('Number of Super Over Matches', color='red')
plt.grid(True)


# ====================== 4th Figure ======================
bat_decision = (
    ipl[ipl['toss_decision'] == 'Batting']
    .groupby('season')['id']
    .count()
)

field_decision = (
    ipl[ipl['toss_decision'] == 'Fielding']
    .groupby('season')['id']
    .count()
)

plt.subplot(3,2,4)

plt.plot(bat_decision.index,
         bat_decision.values,
         marker='o',
         label='Bat First')

plt.plot(field_decision.index,
         field_decision.values,
         marker='o',
         label='Field First')

plt.title('Toss Decision Trend Over Seasons')
plt.xlabel('Season')
plt.ylabel('Count')
plt.legend()
plt.grid(True)


# ====================== 5th Figure ======================
team_name = 'Mumbai Indians'

team_wins = (
    ipl[ipl['winner'] == team_name]
    .groupby('season')['id']
    .count()
)

plt.subplot(3,2,5)

plt.plot(team_wins.index,
         team_wins.values,
         marker='o',
         markerfacecolor='red')

plt.title(f'{team_name} Wins Per Season')
plt.xlabel('Season')
plt.ylabel('Number of Wins')
plt.grid(True)


# ====================== Layout ======================
plt.tight_layout()
plt.show()
```
