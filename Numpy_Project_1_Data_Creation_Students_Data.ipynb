##### **Step 1 :** Import Library 

```python
import numpy as np
```

##### **Step 2 :** Create Dataset
##### Example : Suppose we have marks of 5 students in 4 subjects.Rows = Students, Columns = Subjects

```python
students = np.array([
    [85, 78, 92, 88],
    [76, 81, 79, 85],
    [90, 95, 93, 97],
    [65, 70, 68, 72],
    [88, 84, 91, 89]
])
```

```python
students
```

```python
print(students)
```

##### **Step 3 :** Check Dataset Information

```python
print(students.shape)
```

##### Total 5 Students, 4 Subjects

```python
print(students.ndim)
```

```python
print(students.size)
```

```python
print(students.dtype)
```

##### **Step 4 :** Calculate Total Marks
##### Student Wise Total

```python
total_marks = np.sum(students, axis=1)
print(total_marks)
```

##### **Step 5 :** Calculate Average Marks

```python
average_marks = np.mean(students, axis=1)
print(average_marks)
```

##### **Step 6 :** Find Highest Marks

```python
highest = np.max(students)
print(highest)
```

##### **Step 7 :** Find Lowest Marks

```python
lowest = np.min(students)
print(lowest)
```

##### **Step 8 :** Subject Wise Average

```python
subject_avg = np.mean(students, axis=0)
print(subject_avg)
```

##### **Step 9 :** Find Topper

```python
topper_index = np.argmax(total_marks)
print(topper_index)
```

##### **Step 10 :** Find Weakest Student

```python
weak_student = np.argmin(total_marks)
print(weak_student)
```

##### **Step 11 :** Students Scoring Above 85 Average

```python
average_marks = np.mean(students, axis=1)
result = average_marks > 85
print(result)
```

##### Display Those Students

```python
print(students[average_marks > 85])
```

##### **Step 12 :** Add Bonus Marks
##### Add 5 marks to every subject.

```python
bonus = students + 5
print(bonus)
```

##### **Step 13 :** Grade Assignment

```python
average_marks = np.mean(students, axis=1)
grades = np.where(
    average_marks >= 90, 'A',
    np.where(
        average_marks >= 80, 'B',
        np.where(
            average_marks >= 70, 'C', 'D'
        )
    )
)
print(grades)
```

##### **Step 14 :** Sort Students by Total Marks

```python
sorted_students = np.argsort(total_marks)
print(sorted_students)
```

##### Descending Order

```python
print(np.argsort(total_marks)[::-1])
```

##### **Step 15 :** Standard Deviation

```python
std = np.std(students)
print(std)
```

##### **Step 16 :** Variance

```python
var = np.var(students)
print(var)
```

##### **Step 17 :** Median

```python
median = np.median(students)
print(median)
```
